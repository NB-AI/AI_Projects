The cumulant-generating function $K(t)$ of a random variable $X$ is defined as 
the natural logarithm of the moment-generating function
\begin{equation}
  \newcommand\E{\operatorname{\mathbb{E}}}
  \newcommand\dif{\mathrm{d}}
  K(t) = \log \E\left[e^{tX}\right]. 
\end{equation}
Let $\kappa_n = K^{(n)}(0)$ the $n$-th derivative of $K(t)$ evaluated at $t=0$, 
then the Maclaurin series is 
\begin{equation}
  K(t) = \sum_{n=1}^\infty \kappa_n \frac{t^n}{n!}. 
\end{equation}

<h2 style="color:rgb(0,120,170)">Task 1</h2>
Show that if the random variables $X$ and $Y$ are independent then 
$\kappa_n(X+Y) = \kappa_n(X) + \kappa_n(Y)$ and $\kappa_n(\alpha X) = \alpha^n \kappa_n(X)$
holds.

\begin{align*}

Consider a standard normal random variable $X \sim p(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}}$. Show that $\E[X] = 0$ and $\E[X^n] = (n-1) \E[X^{n-2}]$ for $n \geq 2$ and use this to determine the third and fourth cumulants of $X$. *Hint:* Use the fact that $\int f'(x) e^{f(x)} \dif x = e^{f(x)}$. 

########## YOUR SOLUTION HERE ##########
\begin{align*}
\E(X) &= \frac{1}{\sqrt{2\pi}} \int x \cdot e^{\frac{-x^2}{2}}dx
= \frac{-1}{\sqrt{2\pi}} \int -x \cdot e^{\frac{-x^2}{2}}dx = 
\frac{-1}{\sqrt{2\pi}} \left[e^{-\frac{x^2}{2}}\right]_{-\infty}^{\infty}
= \frac{-1}{\sqrt{2\pi}} \cdot (0 - 0) = 0\\\\
\end{align*}

\begin{align*}
\E(X^n) = \frac{1}{\sqrt{2\pi}} \int x^n e^{\frac{-x^2}{2}} dx
= \frac{1}{\sqrt{2\pi}} \int x^{n-1} \left(x\cdot e^{\frac{-x^2}{2}}\right) dx\\
\end{align*}

Use partial integration: 
\begin{align*}
&\int f' \cdot g = [fg] - \int f\cdot g'\\\\
&f' = x e^{\frac{-x^2}{2}} \\
&f = -e^{\frac{-x^2}{2}}\\
&g = x^{n-1}\\
&g' = (n-1) x^{n-2}\\\\
&\frac{1}{\sqrt{2\pi}} \int x^{n-1} \left(x\cdot e^{\frac{-x^2}{2}}\right) dx\\
&= \frac{1}{\sqrt{2\pi}} \left( \left[x^{n-1} \left(-e^{\frac{-x^2}{2}} \right)  \right]_{-\infty}^{\infty} - \int -e^{\frac{-x^2}{2}} (n-1) x^{n-2} dx    \right) \\
&=  \frac{1}{\sqrt{2\pi}} \left( 0 + (n-1) \int e^{\frac{-x^2}{2}} x^{n-2} dx    \right) \\
&=  \frac{n-1}{\sqrt{2\pi}} \int e^{\frac{-x^2}{2}} x^{n-2} dx    \\
&= (n-1) \E(X^{n-2})
\end{align*}

Use this to determine the third and fourth cumulants of $X$:
\begin{align*}
&\kappa_1 = \E(X) = 0\\
&\kappa_2 = \E(X^2)=(n-1)\E(X^{n-2}) = (2-1) \E(X^{2-2}) = 1 \cdot \E(X^0) = 1 \cdot 1 = 1\\
&\kappa_3 = E(X^3) = (3-1)\E(X^{3-2}) = 2 \E(X) = 2 \cdot 0 = 0\\
&\kappa_4 = E(X^4) -3(E(X^2))^2 = (4-1)\E(X^{4-2})- 3\cdot 1^2 = 3\E(X^2) -3 = 3 \cdot 1 - 3 = 0
\end{align*}

What differences do you observe and how do they relate to the properties of PCA and ICA? 

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA, FastICA

def scale(s, a=0, b=1):
    return (b - a) * (s - s.min()) / (s.max() - s.min()) + a

def show_image(img):
    plt.imshow(img.reshape(3, 32, 32).transpose((1, 2, 0)))
```

```python
########## YOUR SOLUTION HERE ##########
data = np.load('cifar-10.npy')
```

```python
pca = PCA(n_components=100)
pca.fit(data)
U_T = pca.components_ # rows are pricipal components
```

```python
print('Components by PCA:')
fig = plt.figure(figsize=(53,53))
c = 1
for ind,sample_scale in enumerate(U_T):
    scaled_sample = scale(sample_scale)
    U_T[ind] = scaled_sample
    plt.subplot(10,10,c)
    c += 1
    show_image(U_T[ind])
```

Sol: PCs become more and more complicated because defined by the variances.

```python
ica = FastICA(n_components=100)
ica.fit(data)
```

```python
ica_comp = ica.components_
```

```python
print('Components by ICA:')
fig = plt.figure(figsize=(53,53))
c = 1
for ind,sample_scale in enumerate(ica_comp):
    scaled_sample = scale(sample_scale)
    ica_comp[ind] = scaled_sample
    plt.subplot(10,10,c)
    c += 1
    show_image(ica_comp[ind])
    
```

Sol: Similar complex. No order.

ICA is more concentrated on edges than PCA. It gives us information related to edge detection and colour courses.
Furthermore, we know: Neighbouring pixels are often correlated to each other. The single ICA components therefore contain often only the neighbouring pixels (pattern) such that the other component doesn't have to take care of it. Therefore, the components are independent of each other.<br>
In contrast, PCA looks what the pictures have in common. The first components refer to the colour course of the pictures. Furthermore, they aren't statistically independent from each other. E. g. PC1 seems to be the opposite from PC3.

```python
import scipy.io.wavfile
import sounddevice as sd

########## YOUR SOLUTION HERE ##########
```

```python
samplerate_bird, data_bird = scipy.io.wavfile.read('birds.wav')
samplerate_ocean, data_ocean = scipy.io.wavfile.read('ocean.wav')
samplerate_frog, data_frog = scipy.io.wavfile.read('frogs.wav')
```

```python
# Take the left channels
data_bird_left = data_bird.T[0].T
data_ocean_left = data_ocean[:, 0]
data_frog_left = data_frog[:, 0]
```

```python
# Create a mixing matrix U:
# As I learnt it can be arbitrary.
U = np.asarray([[13,19,6],[27,60,33],[-1,-12,4]])
```

```python
data_bird_left = data_bird_left.reshape((data_bird_left.shape[0],1)).T
data_ocean_left = data_ocean_left.reshape((data_ocean_left.shape[0],1)).T
data_frog_left=data_frog_left.reshape((data_frog_left.shape[0],1)).T

Y = np.concatenate((data_bird_left, data_ocean_left, data_frog_left), axis=0)
```

```python
X = U @ Y
```

```python
mix1 = X[0]
mix2 = X[1]
mix3 = X[2]
```

```python
def player(np_sound):
    return sd.playrec(np_sound,channels=1)
```

```python
player(mix1)
```

```python
player(mix2)
```

```python
player(mix3)
```

```python
try:
    import random
    random.seed(0)
except:
    pass
```

```python
fica = FastICA()
Y_found = fica.fit_transform(X.T)
Y_found = Y_found.T
```

```python
# Normalize/scale found Y:
mean1 = Y_found[0].mean()
std1 = Y_found[0].std()
Y_found1 = (Y_found[0] - mean1)/std1

mean2 = Y_found[1].mean()
std2 = Y_found[1].std()
Y_found2 = (Y_found[1] - mean2)/std2

mean3 = Y_found[2].mean()
std3 = Y_found[2].std()
Y_found3 = (Y_found[2] - mean3)/std3
```

```python
print('Reconstructed ocean sound:')
player(Y_found2) # needed shape: (c,1)
```
